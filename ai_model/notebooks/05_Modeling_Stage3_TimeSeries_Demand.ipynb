# 05. 사용주기 AI 3단계 월별 수요 예측

3단계는 자산별 총수명 예측 모델과 별도로, 월별 고장/처분 수요를 직접 예측하는 모델을 탐색하는 실험이다.

- lag/rolling/계절 피처 조합 비교
- XGBoost, RandomForest, ExtraTrees, GradientBoosting, Ridge 비교
- 시간 순서를 지키는 recursive backtest 검증
- UI 월별 고장 예상 수량 그래프에 사용할 월별 수요 모델 선정

실행 결과는 `ai_model/experiments/runs/*_stage3_monthly_model_search`에 저장된다.

```python
from pathlib import Path
import runpy

def find_project_root(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for path in [current, *current.parents]:
        if (path / "dataset" / "create_data" / "data_ml" / "phase4_training_data.csv").exists():
            return path
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
runpy.run_path(
    str(PROJECT_ROOT / "ai_model" / "experiments" / "scripts" / "run_stage3_monthly_model_search.py"),
    run_name="__main__",
)
```
